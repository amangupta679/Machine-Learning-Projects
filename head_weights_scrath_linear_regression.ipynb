```
 %matplotlib inline
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns
 plt.rcParams['figure.figsize']=(20,10)
 plt.rcParams['text.color']='black'
 plt.rc('xtick',labelsize=20)
 plt.rc('ytick',labelsize=20)
```

```
data=pd.read_csv('https://raw.githubusercontent.com/mubaris/potential-enigma/master/headbrain.csv')
data.shape
```

```
data.head()
```

```
sns.pairplot(data,hue='Gender')
```

```
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
plt.scatter(X,Y)
```

```
mean_x=np.mean(X)
mean_y=np.mean(Y)
```

```
n=len(X)
num=0 
denum=0 
for i in range(n):
  num+=(X[i]-mean_x)*(Y[i]-mean_y)
  denum+=(X[i]-mean_x)**2
b1=num/denum 
b0=mean_y-(b1*mean_x) 
(b1,b0)
```

```
y_pred=b1*X+b0
plt.scatter(X,Y,color='g',label='Scatter Plot')
plt.plot(X,y_pred,label='Regression Line')
plt.title('Human Brain')
plt.xlabel('Head Size')
plt.ylabel('Weight')
plt.legend()
plt.show()
```

```
r_num=np.sum((y_pred-mean_y)**2)
r_denum=np.sum((Y-mean_y)**2)
error=r_num/r_denum
error
```
